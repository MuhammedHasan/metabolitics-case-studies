### All Reaction Distribution

Curving on sampling results may provide memory efficent way of calculating and storing results.

```python
%matplotlib inline

import scipy
from scipy.stats import gamma

# dist_names = ['chi2', 'norm', 'exponnorm', 'expon', 'exponpow', 'rayleigh']

dist_names = ['chi2', 'norm']

for r in ecoli.reactions:
    vec_1000 = list(df_1000[r.id].sort_values())
    plt.hist(vec_1000, histtype='step', bins=10)

    min_max = df_min_max.T[r.id]
    print(r.id)
    print(min_max)
    
    vecs_100 = np.array([df[r.id].sort_values() for df in df_100s]).mean(axis=0)
    plt.hist(vecs_100, histtype='step', bins=10)

    pdf_x = np.linspace(min_max.minimum, min_max.maximum, 1000)
    for dist_name in dist_names:
        
        fit_vec_1000 = vec_1000
        fitted_pdf_x = pdf_x
        
        if min_max.maximum <= 0:
            fit_vec_1000 = np.array(vec_1000) * -1
            fitted_pdf_x = pdf_x * -1
            
        dist = getattr(scipy.stats, dist_name)
        param = dist.fit(fit_vec_1000)
        
        pdf_fitted = dist.pdf(fitted_pdf_x, *param[:-2], loc=param[-2]) * 1000
                
        plt.plot(pdf_x, pdf_fitted)
    
    plt.show()
```
